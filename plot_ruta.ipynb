```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Carregar el dataset

df = pd.read_csv('dades_cotxe_gps_ruta_realtime_20250422_162342.csv')

# Crear el gràfic
plt.figure(figsize=(10, 8))
plt.plot(df['Longitud'], df['Latitud'], marker='o', linestyle='-', linewidth=1, markersize=3)

# Afegir etiquetes i títol
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.title('Ruta')
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box') # Assegura que l'escala sigui la mateixa per a ambdós eixos

# Mostrar el gràfic
plt.show()
```
